# 보스톤 주택 가격 - 다중 회기

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['PRICE'] = boston.target
df.head()
```

# - 산점도와 선형회귀 직선

```python
fig, axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)
features = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']
for i, features in enumerate(features):
    row, col = i // 4, i % 4
    sns.regplot(x=features, y='PRICE',data=df,ax=axs[row][col])
```

# -어떤 feature가 price와 상관관계가 높을까?

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
for features in boston.feature_names:
    X = df[features].values.reshape(-1,1)
    y = boston.target
    lr = LinearRegression()
    lr.fit(X,y)
    r2 = lr.score(X,y)
    print(f'{features}: {r2:.4f}')
```
