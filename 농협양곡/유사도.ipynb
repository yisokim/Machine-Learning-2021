```python
# data load
import pandas as pd
data = pd.read_csv("C:/nh/자연어처리.csv", low_memory=False)
data.head(2)
```

```python
# 결측치 확인
data = data.fillna('')
data.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_matrix = TfidfVectorizer().fit_transform(data['리뷰'])
print(tfidf_matrix.shape)
```

리뷰 칼럼에 대해서 tf-idf를 수행한 결과, 587개의 리뷰를 표현하기 위해 총 3464개의 단어가 사용되었음을 보여주고 있다.
이제 코사인 유사도를 사용하면 바로 문서의 유사도를 구할 수 있음.

```python
# TF-IDF 클래스로 BOW 만들기
tfidfv = TfidfVectorizer().fit(data['리뷰'])
print(tfidfv.transform(data['리뷰']).toarray())
print(tfidfv.vocabulary_)
print(tfidfv.vocabulary_)
```

```python
# 비교를 위해 CountVectorizer로 동일한 코퍼스를 벡터화
from sklearn.feature_extraction.text import CountVectorizer
corpus = data['리뷰']
vector = CountVectorizer()
print(vector.fit_transform(corpus).toarray()) # 코퍼스로부터 각 단어의 빈도 수를 기록한다.
print(vector.vocabulary_) # 각 단어의 인덱스가 어떻게 부여되었는지를 보여준다.
```

```python
# 코사인 유사도 사용
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# 리뷰 d와 인덱스를 가진 테이블 생성. 리뷰 아이디를 입력하면 인덱스를 리턴하기 위함
indices = pd.Series(data.index, index=data['id']).drop_duplicates()
print(indices.head())
```

```python
idx = indices['a1']
print(idx)
```

```python
def get_recommendations(title, cosine_sim = cosine_sim):
    idx = indices[id]
    
    # 모든 영화에 대해서 해당 영화의 유사도를 구한다.
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # 유사도에 따라 영화들을 정렬한다.
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # 가장 유사한 10개의 영화를 받아온다.
    sim_scores = sim_scores[1:11]
    
    # 가장 유사한 10개의 영화의 인덱스를 받아온다.
    kind_indices = [i[0] for i in sim_scores]
    
    # 가장 유사한 10개의 영화의 제목을 리턴한다.
    return data['id'].iloc[kind_indices]
```
